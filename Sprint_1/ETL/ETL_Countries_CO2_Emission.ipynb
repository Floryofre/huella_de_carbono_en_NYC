#### carga de archivo csv

#### se eliminan las columnas que no son necesarias para el estudio del la contaminacion en NYC

```python
# Eliminar las columnas especificadas
df_Countries_CO2_Emission = df_Countries_CO2_Emission.iloc[:, 1:]
```

#### cambia los nombres de las columnas 

```python
# Creamos un diccionario con los nombres de las columnas antiguas como claves y los nuevos nombres como valores
nuevos_nombres = {
    'Country': 'pais',
    'Energy_type': 'tipo_energia',
    'Year': 'anio',
    'Energy_consumption': 'consumo_de_energia',
    'Energy_production': 'produccion_de_energia',
    'GDP': 'PIB',
    'Populaition': 'poblacion',
    'Energy_intensity_per_capita': 'intensidad_energetica_per_capita',
    'Energy_intensity_by_GDP': 'intensidad_energetica_por_PIB',
    'CO2_emission': 'emision_de_CO2'
}

# Renombramos las columnas usando el diccionario creado
df_Countries_CO2_Emission = df_Countries_CO2_Emission.rename(columns=nuevos_nombres)
```

```python
df_Countries_CO2_Emission
```

#### crea el archivo parquet y csv

```python
# Guardar el DataFrame en un archivo CSV llamado 'NYC_air_quality_clean.csv' y en un archivo PARQUET 'NYC_air_quality.parquet'
df_Countries_CO2_Emission.to_csv('Countries_CO2_Emission.csv', index=False)
df_Countries_CO2_Emission.to_parquet('Countries_CO2_Emission.parquet', index=False)
```
